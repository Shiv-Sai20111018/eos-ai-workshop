```python
def calculate_percentage(marks_list):
    # 1. Add all marks
    total = sum(marks_list)

    # 2. Divide by max possible (each out of 100)
    percentage = total / (len(marks_list) * 100) * 100

    # 3. Round to 2 decimal places
    return round(percentage, 2)


# Use it
marks = [88, 92, 76, 85, 90]
result = calculate_percentage(marks)
print(f"Percentage: {result}%")   # Percentage: 86.2%
```

```python
name = "Priya"
score = 92.5678
percentage = 0.867

# Basic
print(f"Hi {name}")              # Hi Priya

# Format decimals (2 places)
print(f"Score: {score:.2f}")     # Score: 92.57

# Format as percentage
print(f"{percentage:.1%}")       # 86.7%

# Align + pad
print(f"{name:>10}")             # "     Priya"  (right-aligned, width 10)

# Math inside braces
print(f"Next year: {score + 5:.1f}")   # Next year: 97.6
```

```python
import numpy as np

scores = np.array([88, 92, 76, 85, 90])

print(scores.mean())    # 86.2      ← average
print(scores.std())     # 5.56      ← standard deviation (spread)
print(scores.max())     # 92
print(scores.min())     # 76
print(scores.sum())     # 431

# Where's the max/min?
print(np.argmax(scores))  # 1       ← index of the top scorer (92)
print(np.argmin(scores))  # 2       ← index of the lowest (76)

# Sort
print(np.sort(scores))    # [76 85 88 90 92]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Create a 16×16 canvas of all zeros (black)
canvas = np.zeros((16, 16), dtype=int)

# 2. Set some cells to 1 (white) to "draw"
canvas[5, 5] = 1        # single pixel at row 5, col 5
canvas[5:8, 10] = 1     # vertical line of 3 pixels
canvas[2, :] = 1        # full top row

# 3. Use slicing to add a border
canvas[0, :] = 1
canvas[-1, :] = 1
canvas[:, 0] = 1
canvas[:, -1] = 1

# 4. Display
plt.imshow(canvas, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 4096x4096 canvas
canvas = np.zeros((4096,4096), dtype=int)

# Border
canvas[0,:] = 1
canvas[-1,:] = 1
canvas[:,0] = 1
canvas[:,-1] = 1

# Eyes
canvas[800:1000, 1000:1200] = 1     # left eye
canvas[800:1000, 2800:3000] = 1     # right eye

# Eyebrows
canvas[700:750, 950:1250] = 1       # left eyebrow
canvas[700:750, 2750:3050] = 1      # right eyebrow

# Nose
canvas[1600:2000, 1900:2100] = 1

# Mouth (smile curve)
canvas[2800:2850, 1200:2800] = 1
canvas[2851:2900, 1300:2700] = 1
canvas[2901:2950, 1400:2600] = 1

# Forehead mark
canvas[500:600, 1900:2100] = 1

# Hair outline
canvas[200:600, 800:3200] = 1

# Display
plt.imshow(canvas, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create 8192x8192 canvas
canvas = np.zeros((8192,8192), dtype=int)

# Generate indices grid
Y, X = np.indices(canvas.shape)

# Head circle
center_x, center_y, radius = 4096, 4096, 3000
mask_head = (X-center_x)**2 + (Y-center_y)**2 < radius**2
canvas[mask_head] = 1

# Eyes (two smaller circles)
mask_left_eye = (X-3000)**2 + (Y-3000)**2 < 300**2
mask_right_eye = (X-5200)**2 + (Y-3000)**2 < 300**2
canvas[mask_left_eye] = 0   # black eyes
canvas[mask_right_eye] = 0

# Nose (small circle)
mask_nose = (X-4096)**2 + (Y-4200)**2 < 200**2
canvas[mask_nose] = 0

# Mouth (arc of a circle)
mask_mouth = ((X-4096)**2 + (Y-5000)**2 < 1200**2) & (Y>5000)
canvas[mask_mouth] = 0

# Display
plt.figure(figsize=(10,10))
plt.imshow(canvas, cmap='gray')
plt.axis('off')
plt.show()
```
