```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# 1. Make the background (light blue sky)
canvas = np.zeros((64, 64), dtype=int)

# 2. Cat body (orange oval)
canvas[20:44, 20:44] = 1

# 3. Cat head (orange square-ish block above body)
canvas[10:20, 24:40] = 1

# 4. Cat ears (two triangles on top of head)
canvas[8:12, 24:28] = 1
canvas[8:12, 36:40] = 1

# 5. Eyes (white pixels)
canvas[14:16, 28:30] = 2
canvas[14:16, 34:36] = 2

# 6. Nose (pink pixel)
canvas[17, 31:33] = 3

# 7. Tail (orange vertical line on right side)
canvas[24:44, 46:48] = 1

# 8. Stadium border (gray outline)
canvas[0, :] = 4
canvas[-1, :] = 4
canvas[:, 0] = 4
canvas[:, -1] = 4

# 9. Colors mapped
cmap = ListedColormap([
    "skyblue",    # 0 = background
    "orange",     # 1 = cat body/head/ears/tail
    "white",      # 2 = eyes
    "pink",       # 3 = nose
    "gray"        # 4 = border
])

# 10. Show it
plt.imshow(canvas, cmap=cmap)
plt.axis('off')
plt.show()
```
